To check the missing values

```python
df.isnull().sum()
```

```python
numeric_columns = ['Avg VTAT','Avg CTAT','Booking Value','Ride Distance','Driver Ratings','Customer Rating']
df.fillna(df[numeric_columns].mean(),inplace=True)
```

```python
categorical_columns = ['Reason for cancelling by Customer','Driver Cancellation Reason','Incomplete Rides Reason']
df[categorical_columns]=df[categorical_columns].fillna("Unknown")
```

Remove Duplicates

```python

df.drop_duplicates(inplace=True)
```

Convert Date Column

```python

df['Date'] = pd.to_datetime(df['Date'])
```

Extract Hour

```python
df['Hour'] = pd.to_datetime(df['Time']).dt.hour
```

Extract Day Name

```python
df['Day'] = df['Date'].dt.day_name()
```

Extract Month

```python
df['Month'] = df['Date'].dt.month_name()
```

Total Bookings

```python
total_bookings = df['Booking ID'].count()

total_bookings
```

Total Revenue

```python
total_revenue = df['Booking Value'].sum()

total_revenue
```

Average Ride Value

```python
avg_booking = df['Booking Value'].mean()

avg_booking
```

Peak Booking Hours

```python
peak_hours = df.groupby('Hour')['Booking ID'].count()

peak_hours
```

 Vehicle Demand

```python
vehicle_demand = df['Vehicle Type'].value_counts()

vehicle_demand
```

Revenue by Vehicle

```python
vehicle_revenue = df.groupby('Vehicle Type')['Booking Value'].sum()

vehicle_revenue
```

Booking Status Analysis

```python
booking_status = df['Booking Status'].value_counts()
booking_status
```

Customer Cancellation Reasons

Most Popular Routes

```python
routes = df.groupby(['Pickup Location','Drop Location'])['Booking ID'].count().sort_values(ascending=False)
routes.head(10)
```

Peak Hour Chart

```python
peak_hours.plot(kind='bar')

plt.title('Peak Booking Hours')
plt.xlabel('Hour')
plt.ylabel('Bookings')
plt.show()
```

Revenue by Vehicle

```python
vehicle_revenue.plot(kind='bar')
plt.title('Revenue by Vehicle Type')
plt.show()
```
